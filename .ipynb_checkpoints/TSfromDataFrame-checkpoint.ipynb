```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dat = pd.DataFrame.from_csv('./DMResults_g_lcoe.txt', index_col=None, sep=' ')
```

```python
str_times = [str(x) for x in range(86)]
melt_dat = pd.melt(dat, id_vars=['reactor_build_decision', 'htgr_capital_cost', 'sfr_capital_cost'], value_vars=str_times, var_name='Time Step', value_name='Name of Number')
new_dat = melt_dat.drop(['htgr_capital_cost', 'sfr_capital_cost'], axis=1)
```

```python
new_dat['Time Step'] = new_dat['Time Step'].astype(float)
```

```python
sns.tsplot(data=new_dat, time='Time Step', condition='reactor_build_decision', value='Name of Number', n_boot=False, estimator=np.mean) 
plt.show()
```

```python
# check for any nan or non-floats
print any([np.isnan(x) for x in new_dat['Name of Number']])
print all([isinstance(x, float) for x in new_dat['Name of Number']])
```

```python
sns.set_style('ticks')
sns.set_context('talk')
clrs = sns.hls_palette(4)
for i in range(4):
    sns.regplot(data=new_dat[new_dat['reactor_build_decision'] == i], x='Time Step', y='Name of Number', x_estimator=np.median, x_ci=68, fit_reg=False, color=clrs[i]) 
```
