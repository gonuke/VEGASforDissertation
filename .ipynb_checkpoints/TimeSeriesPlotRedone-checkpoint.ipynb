```python
time_points_str = [str(x) for x in range(86)]
time_points = np.array(range(86))
```

```python
min_values = dat_min[time_points_str].values.T
max_values = dat_max[time_points_str].values.T
med_values = dat_median[time_points_str].values.T
```

```python
min_values.shape
```

```python
time_points.shape
```

```python
RBD = [0]*86 + [1]*86 + [2]*86 + [3]*86
RBD = np.array(RBD)
```

```python
min_values[:10]
```

```python
# order is fortran convention to stack each column on top of each other
# could also do some kind of concantenate with a list of 4 elements, each being a column
min_values = np.reshape(min_values, 86*4, order='F')
max_values = np.reshape(max_values, 86*4, order='F')
med_values = np.reshape(med_values, 86*4, order='F')
```

```python
min_values[:10]
```

```python
long_time = np.concatenate((time_points, time_points, time_points, time_points))
```

```python
matrix_data = np.array([long_time, min_values, med_values, max_values, RBD]).T
```

```python
matrix_data
```

```python
new_dat = pd.DataFrame(matrix_data, columns=['Time Step', 'Minimum', 'Median', 'Maximum', 'Reactor Build Decision'])
```

```python
new_dat
```

```python
sns.tsplot(data=new_dat, time='Time Step', condition='Reactor Build Decision', value='Median', err_style=None, unit=None, ci=None, )
```

```python
new_dat.head(10)
```

```python
sns.set_style('ticks')
sns.set_context('talk')
sns.lmplot(x='Time Step', y='Median', data=new_dat, hue='Reactor Build Decision', fit_reg=False)
plt.show()
```
